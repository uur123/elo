```
from google.colab import files
uploaded1 = files.upload()
```

```
df_sr = pd.read_csv(io.BytesIO(uploaded1['Sample Request (8).csv']))
```

```
df_tsr[['RequestDate', 'DueDate', 'StartDate']] = df_tsr[['RequestDate', 'DueDate', 'StartDate']].apply(pd.to_datetime)
df_tsr_2023 = df_tsr[df_tsr['RequestDate'].dt.year > 2022]
df_tsr_2023 = df_tsr_2023.dropna(subset=['TSRId'])
```

```
df_sr_1 = df_sr[df_sr['TSRID'] >= 640]
df_sr_cleaned = df_sr_1.dropna(subset=['TSRID'])
df_sr_cleaned['TSRID'] = df_sr_cleaned['TSRID'].astype(str)
df_sr_cleaned['TSRID'] = 'FDSR00' + df_sr_cleaned['TSRID']
df_sr_cleaned = df_sr_cleaned.rename(columns = {'TSRID':'TSRId'})
df_sr_cleaned['TSRId'] = df_sr_cleaned['TSRId'].str.removesuffix(".0")
```

```
mergedRes = pd.merge(df_sr_cleaned, df_tsr_2023, on ='TSRId')
mergedRes = mergedRes.drop(columns=['StartDate_x', 'Created By_x','Assigned to', 'DueDate_y','Comments_y', 'Location: State','Location: City', 'AssignedTo_y', 'Created By_y', 'Attachments','StartDate_y',  'Label setting'])
       
```

```
mergedRes.to_csv('mergedTSR.csv', encoding = 'utf-8-sig') 
files.download('mergedTSR.csv')
```
